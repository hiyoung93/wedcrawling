```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
i = 1
url = 'https://www.instagram.com/explore/tags/doll/'

def crawl(url):
    data = requests.get(url)
    print(data, url)
    return data.content
pageString = crawl(url)
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(5)
driver.get(url)
log = driver.find_element_by_xpath("//div[@class='v1Nh3 kIKUG  _bz0w']").click()
time.sleep(5)
d = driver.current_url
pageString = crawl(d)
bsOj = BeautifulSoup(pageString, 'html.parser')
s = bsOj.find('title')
f = open('dump2.txt', 'a', -1,'utf-8')
print(s, file=f)
f.close()
for  i in range(1,10001) : 
    time.sleep(5)
    log = driver.find_element_by_xpath("//*[@class='HBoOv coreSpriteRightPaginationArrow']").click()
    time.sleep(5)
    d = driver.current_url
    pageString = crawl(d)
    bsOj = BeautifulSoup(pageString, 'html.parser')
    s = bsOj.find('title')
    f = open('dump2.txt', 'a', -1,'utf-8')
    print(s, file=f)
    f.close()
    print(i+1)
```
